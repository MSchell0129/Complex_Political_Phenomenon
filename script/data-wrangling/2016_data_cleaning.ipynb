```python
import sqlite3
import pandas as pd
```

```python
#Connecting to the database sqlite
database_path = "../election_resources/Elections_db.sqlite"
conn = sqlite3.connect(database_path)
query = "SELECT * FROM presidential_polls_2016"
df = pd.read_sql_query(query, conn)

conn.close()
```

```python
df.head()
```

```python
df['state'].value_counts()
```

```python
new_df = df.loc[df['state'] == 'U.S.'] 
new_df.head()
```

```python
new_df['state'].value_counts()
```

```python
columns_to_drop =['id', 'pollster', 'grade', 'rawpoll_clinton', 'rawpoll_trump', 'rawpoll_johnson', 'rawpoll_mcmullin', 'url', 'question_id', 'createddate', 'timestamp', 'cycle', 
                  'branch', 'type', 'matchup', 'forecastdate', 'state', 'startdate', 'pollster', 'grade', 'adjpoll_johnson', 'adjpoll_mcmullin', 'multiversions', 'poll_id', 'population', 'poll_wt']
df = new_df.drop(columns=columns_to_drop)
```

```python
df.head()
```

```python
r = df.copy()
d = df.copy()
```

```python
r['party'] = 'r'
d['party'] = 'd'
```

```python
r['favorability'] = r['adjpoll_trump']
d['favorability'] = d['adjpoll_clinton']
r = r.drop(columns=['adjpoll_clinton', 'adjpoll_trump'])
d = d.drop(columns=['adjpoll_trump', 'adjpoll_clinton'])
```

```python
r.head()
```

```python
rd2_df = pd.concat([r, d], ignore_index=True)
rd2_df.head()
```

```python
rd2_df.tail()
```

```python
rd2_df['favorability'] = rd2_df['favorability'] - 50
```

```python
results = {'d': 0, 'r': 1}
rd2_df['results'] = rd2_df['party'].apply(lambda x: results.get(x))
rd2_df.head()
```

```python
r = rd2_df.loc[rd2_df['party'] == 'r']
d = rd2_df.loc[rd2_df['party'] == 'd']
display(r.head())
display(d.head())
```

```python
r2d2 = r.merge(d, on= ['enddate', 'samplesize'], how = 'left')
r2d2.head()
r2d2.tail()
```

```python
r2d2.isna().sum()
```

```python
r2d2.shape
```

```python
r2d2 = r2d2.rename(columns={'favorability_x': 'favorability_r', 'favorability_y': 'favorability_d', 'results_x': 'results'})
r2d2 = r2d2.drop(columns=['results_y', 'party_y', 'party_x'])
display(r2d2.head())
display(r2d2.tail())
```

```python
r2d2.head()
```

```python
r2d2.to_csv('../election_resources/cleaned_presidential_polls_2016.csv', index=False)
```
