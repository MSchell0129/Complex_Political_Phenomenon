```python
import pandas as pd
```

```python
polling_df = pd.read_csv('../election_resources/favorability_polls_rv_2019.csv')
polling_df.head()
```

```python
polling_df.tail()
```

```python
polling_df.describe()
```

```python
polling_df.info()
```

```python
polling_df.drop(columns=['start_date', 'pollster_id', 'sponsors', 'methodology', 'url'], inplace=True)
polling_df.head()
```

```python
end_date = pd.to_datetime(polling_df['end_date'], format='%m/%d/%Y')
```

```python
polling_df['end_date'] = end_date
```

```python
polling_df.info()
```

```python
trump_df = polling_df.loc[polling_df['politician'] == 'Donald Trump']
trump_df
```

```python
polling_df.head()
```

```python
biden_df = polling_df.loc[polling_df['politician'] == 'Joseph R. Biden Jr.']
biden_df
```

```python
biden_v_trump = pd.concat([biden_df, trump_df], ignore_index=True)
biden_v_trump.head()
```

```python
results = {'Joseph R. Biden Jr.': 1, 'Donald Trump': 0}
```

```python
biden_v_trump['results'] = biden_v_trump['politician'].apply(lambda x: results.get(x))
biden_v_trump.head()
```

```python
biden_v_trump.fillna(0, inplace=True)
```

```python
biden_v_trump.to_csv('../election_resources/biden_v_trump.csv', index=False)
biden_v_trump.to_json('../election_resources/biden_v_trump.json', orient='records')
```
