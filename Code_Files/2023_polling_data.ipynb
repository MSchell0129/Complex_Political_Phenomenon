```python
results = {'Donald Trump': 0, 'Joe Biden': 0, 'Mike Pence': 0,
       'Ron DeSantis': 0, 'Vivek G. Ramaswamy': 0, 'Tim Scott': 0,
       'Robert F. Kennedy': 0, 'Nikki Haley': 0, 'Chris Christie': 0,
       'Marianne Williamson': 0, 'Asa Hutchinson': 0, 'Doug Burgum': 0,
       'Larry A. Elder': 0, 'Francis Suarez': 0, 'Will Hurd': 0}
```

```python
polling_df['results'] = polling_df['politician'].apply(lambda x: results.get(x))
polling_df.head()
```

```python

desired_columns = ['question_id', 'end_date', 'pollster', 'sample_size', 'population', 'politician',
                   'favorable', 'unfavorable', 'very_favorable', 'somewhat_favorable',
                   'somewhat_unfavorable', 'very_unfavorable', 'results']

# Reorder the columns in the DataFrame
polling_df = polling_df[desired_columns]

# Print the modified DataFrame
polling_df
```

```python
trump_df = polling_df.loc[polling_df['politician'] == 'Donald Trump']
trump_df
```

```python
biden_df = polling_df.loc[polling_df['politician'] == 'Joe Biden']
biden_df
```

```python
biden_v_trump = pd.concat([biden_df, trump_df], ignore_index=True)
```

```python
biden_v_trump.fillna(0, inplace=True)
```

```python
favorable = biden_v_trump['favorable']*0.6667
unfavorable = biden_v_trump['unfavorable']*-0.6667
somewhat_favorable = biden_v_trump['somewhat_favorable']*0.3333
somewhat_unfavorable = biden_v_trump['somewhat_unfavorable']*-0.3333
very_favorable = biden_v_trump['very_favorable']*1
very_unfavorable = biden_v_trump['very_unfavorable']*-1
biden_v_trump['favorability'] = favorable + unfavorable + somewhat_favorable + somewhat_unfavorable + very_favorable + very_unfavorable
```

```python
biden_v_trump.drop(columns=['question_id', 'end_date', 'pollster', 'population', 'favorable', 'unfavorable', 'very_favorable','somewhat_favorable', 'somewhat_unfavorable', 'very_unfavorable',], inplace=True)
biden_v_trump.head()
```

```python
party = {'Joe Biden': 'd', 'Donald Trump': 'r'}
biden_v_trump['politician'] = biden_v_trump['politician'].map(party)
biden_v_trump.head()
```

```python
biden_v_trump.rename(columns={'politician': 'party', 'sample_size': 'samplesize'}, inplace=True)
biden_v_trump.head()
```

```python
biden_v_trump.to_csv('../election_resources/biden_v_trump24.csv', index=False)
biden_v_trump.to_json('../election_resources/biden_v_trump24.json', orient='records')
```
