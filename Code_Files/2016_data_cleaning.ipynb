```python
import pandas as pd
```

```python
data = pd.read_csv('../election_resources/presidential_polls_2016.csv')
df= pd.DataFrame(data)
```

```python
columns_to_drop =['pollster', 'grade', 'rawpoll_clinton', 'rawpoll_trump', 'rawpoll_johnson', 'rawpoll_mcmullin', 'url', 'question_id', 'createddate', 'timestamp']
df = df.drop(columns=columns_to_drop)
```

```python
df = df.fillna(0)
df['multiversions'] = pd.to_numeric(df['multiversions'], errors='coerce')
df['multiversions'] = df['multiversions'].astype(float)
```

```python
abb = {'lv': 'likely voters', 'rv': 'registered voters', 'a': 'adults', 'v': 'voters'}
df['population']= df['population'].replace(abb)
df.head()
```

```python
df.to_csv('../election_resources/cleaned_presidential_polls_2016.csv', index=False)
```
