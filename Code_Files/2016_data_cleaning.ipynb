```python
import pandas as pd
```

```python
data = pd.read_csv('../election_resources/presidential_polls_2016.csv')
df= pd.DataFrame(data)
df.head()
```

```python
df['state'].value_counts()
```

```python
new_df = df.loc[df['state'] == 'U.S.'] 
new_df.head()
```

```python
new_df['state'].value_counts()
```

```python
columns_to_drop =['pollster', 'grade', 'rawpoll_clinton', 'rawpoll_trump', 'rawpoll_johnson', 'rawpoll_mcmullin', 'url', 'question_id', 'createddate', 'timestamp', 'cycle', 
                  'branch', 'type', 'matchup', 'forecastdate', 'state', 'startdate', 'pollster', 'grade', 'adjpoll_johnson', 'adjpoll_mcmullin', 'multiversions', 'poll_id', 'population', 'poll_wt']
df = new_df.drop(columns=columns_to_drop)
```

```python
df.head()
```

```python
r = df.copy()
d = df.copy()
```

```python
r['party'] = 'r'
d['party'] = 'd'
```

```python
r['favorability'] = r['adjpoll_trump']
d['favorability'] = d['adjpoll_clinton']
r = r.drop(columns=['adjpoll_clinton', 'adjpoll_trump'])
d = d.drop(columns=['adjpoll_trump', 'adjpoll_clinton'])
```

```python
r.head()
```

```python
rd2_df = pd.concat([r, d], ignore_index=True)
rd2_df.head()
```

```python
rd2_df.tail()
```

```python
rd2_df['favorability'] = rd2_df['favorability'] - 50
```

```python
rd2_df.drop(columns=['enddate'], inplace=True)
```

```python
rd2_df.head()
```

```python
results = {'d': 0, 'r': 1}
rd2_df['results'] = rd2_df['party'].apply(lambda x: results.get(x))
rd2_df.head()
```

```python
rd2_df.shape
```

```python
rd2_df.to_csv('../election_resources/cleaned_presidential_polls_2016.csv', index=False)
```
