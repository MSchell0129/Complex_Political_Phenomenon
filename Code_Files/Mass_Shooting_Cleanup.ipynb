```python
import pandas as pd
```

```python
df_2014 = pd.read_csv('../Resources/Mass_Shootings_2014.csv')
df_2015 = pd.read_csv('../Resources/Mass_Shootings_2015.csv')
df_2016 = pd.read_csv('../Resources/Mass_Shootings_2016.csv')
df_2017 = pd.read_csv('../Resources/Mass_Shootings_2017.csv')
df_2018 = pd.read_csv('../Resources/Mass_Shootings_2018.csv')
df_2019 = pd.read_csv('../Resources/Mass_Shootings_2019.csv')
df_20_23 = pd.read_csv('../Resources/Mass_Shootings_20-23.csv')
```

```python
mass_shootings = pd.concat([df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_20_23])
mass_shootings.head()
```

```python
mass_shootings.drop(columns='Operations', inplace=True)
mass_shootings.head()
```

```python
mass_shootings.to_csv('../Resources/mass_shootings_cleaned.csv', index=False)
```
