```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
polling_df.drop(columns=['pollster', 'grade', 'rawpoll_clinton', 'rawpoll_trump', 'rawpoll_johnson',
       'rawpoll_mcmullin', 'url', 'question_id', 'createddate', 'timestamp'], inplace=True)
```

lv = likely voter
rv = registered voters
a = adult
v = voters

```python
polling_df['population'].unique()
```

```python
polling_df = polling_df.fillna(0)
```

```python
polling_df.dtypes
```

```python
polling_df.head()
```

```python
abb = {'lv': 'likely voters', 'rv': 'registered voters', 'a': 'adults', 'v': 'voters'}
```

```python
polling_df['population']= polling_df['population'].replace(abb)
polling_df.head()
```

```python
polling_df['multiversions'] = pd.to_numeric(polling_df['multiversions'], errors='coerce')
polling_df['multiversions'] = polling_df['multiversions'].astype(float)
```

```python
polling_df.dtypes
```

```python
polling_df.to_csv('../election_resources/cleaned_2016_polls.csv')
```
