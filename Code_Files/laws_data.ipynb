```python
import pandas as pd
import json
import requests
import numpy as np
import matplotlib.pyplot as plt
from api_key import laws_api_key
```

```python
offset = 0
limit = 250
laws = []
while offset < 100000:
    offset += 250
    url = f'https://api.congress.gov/v3/bill?offset={offset}&limit=250&fromDateTime=2013-08-09T00%3A00%3A00Z&api_key={laws_api_key}'
    response = requests.get(url).json()
    laws+=response['bills']

```

```python
df_laws = pd.DataFrame(laws)
df_laws.tail(20)
```

```python
display(len(df_laws))
display(df_laws['number'].nunique())
```

```python
df_laws.to_csv('../Resources/Laws/uncleaned_laws.csv', index=False)
```
