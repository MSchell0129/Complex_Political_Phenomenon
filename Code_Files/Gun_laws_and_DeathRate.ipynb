```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import balanced_accuracy_score, confusion_matrix, classification_report
```

## Import data and create data frames

```python
state_gun_laws_df = pd.read_csv('../Resources/Crime_Data/state_gun_law_state.csv')
state_gun_laws_df.head(30)
```

```python
cdc_death_rate_df= pd.read_csv('../Resources/Crime_Data/CDC_death_rate_state20.csv')
cdc_death_rate_df.head()
```

```python
cdc_death_rate_df = cdc_death_rate_df.drop(columns=['URL'])
```

```python
cdc_death_rate_df
```

```python
state_law_and_death_rate=state_gun_laws_df.merge(cdc_death_rate_df, left_on=['state', 'year'], right_on=['STATE', 'YEAR'], how='left')
state_law_and_death_rate
```

```python
state_law_and_death_rate=state_law_and_death_rate.dropna()
state_law_and_death_rate
```

```python
clean_statelaw_deathrate_df= state_law_and_death_rate.drop(columns=["STATE","YEAR"])
clean_statelaw_deathrate_df
```

```python
clean_statelaw_deathrate_df.rename(columns={"RATE": "Rate_in_100k", "DEATHS": "Total_Deaths"}, inplace=True)
clean_statelaw_deathrate_df
```

```python
clean_statelaw_deathrate_df.to_csv('../Resources/Crime_Data/state_gun_and_deathrate.csv', index=False)
```
